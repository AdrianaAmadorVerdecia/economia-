```python
import json
with open("data/mipymes.json","r", encoding="utf-8") as f:
    mipymes = json.load(f)
print(json.dumps(mipymes, indent=2))

print("cantidad de productos:", len(mipymes))
```

### El programa recorre la lista de MIPYMES, cuenta cuántos productos tiene disponibles cada una, ordena los resultados de mayor a menor y finalmente muestra en pantalla qué MIPYME ofrece más variedad de productos.

```python
productos_por_mipyme = []

for mipyme in mipymes["mipymes"]:
    contador = 0
    productos = mipyme["productos"]

    for producto in productos.values():
        if producto["disponible"] == True:
            contador += 1

    productos_por_mipyme.append((mipyme["mipyme_id"], contador))


# Ordenamos de mayor a menor
productos_por_mipyme.sort(key=lambda x: x[1], reverse=True)


# Mostramos ordenado
print("Productos disponibles por MIPYME (ordenado)")
for id_mipyme, cantidad in productos_por_mipyme:
    print("MIPYME", id_mipyme, "->", cantidad, "productos")
```

### La grafica muestra la cant de productos disponibles por mipymes

```python
import matplotlib.pyplot as plt

# Extraemos IDs y cantidades de la lista
ids = [f"MIPYME {id_mipyme}" for id_mipyme, cantidad in productos_por_mipyme]
cantidades = [cantidad for id_mipyme, cantidad in productos_por_mipyme]

# Crear la figura
plt.figure(figsize=(12,8))
bars = plt.barh(ids, cantidades, color="#4CAF50")

# Títulos y etiquetas
plt.xlabel("Cantidad de productos disponibles")
plt.ylabel("MIPYMES")
plt.title("Productos disponibles por MIPYME")

# Añadir el número de productos al final de cada barra
for bar, cantidad in zip(bars, cantidades):
    plt.text(cantidad + 0.2, bar.get_y() + bar.get_height()/2,
             str(cantidad),
             va="center", ha="left", fontsize=10, color="black")

# Ajustar diseño
plt.tight_layout()
plt.show()
```

## En el siguiente analicis se estara presentado el precio promedio de cada producto y en que mipme se puede encontrar mas barato

```python
precios_por_producto = {}
info = mipymes["mipymes"]

for m in info:
    id_mipyme = m["mipyme_id"]
    nombre_mipyme = m["nombre"]
    productos = m["productos"]

    for id_prod, p in productos.items():
        if p["disponible"] == True and p["precio"] > 0:

            if id_prod not in precios_por_producto:
                precios_por_producto[id_prod] = []

            precios_por_producto[id_prod].append(
                (p["precio"], nombre_mipyme,  id_mipyme)
            )

print("Producto | Promedio | Mipyme más barata(nombre + id)")

for id_prod, lista in precios_por_producto.items():
    suma = 0
    for precio, _, _ in lista:
        suma += precio

    promedio = suma / len(lista)

    mas_barata = min(lista, key=lambda x: x[0])

    print(
        "Producto", id_prod,
        "| Prom:", round(promedio, 2),
        "| Más barata:", mas_barata[1],
        "|id:", mas_barata[2],
        "(precio:", mas_barata[0], ")"
    )
```

### El gráfico muestra el precio promedio de cada producto ofrecido por las MIPYMES, permitiendo comparar de manera ordenada cómo varían los valores entre los distintos productos.

```python
import matplotlib.pyplot as plt

productos = []
promedios = []

for id_prod, lista in precios_por_producto.items():
    # Calcular promedio de precios por producto
    promedio = sum(precio for precio, _, _ in lista) / len(lista)
    productos.append(id_prod)
    promedios.append(promedio)

plt.figure(figsize=(10,6))
plt.plot(productos, promedios, marker="o", linestyle="-", color="blue", label="Precio promedio")

plt.xlabel("Productos")
plt.ylabel("Precio en CUP")
plt.title("Precio promedio por producto en MIPYMES")
plt.xticks(rotation=45, ha="right")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import json
with open("productos_revolico.json", "r", encoding="utf-8") as f:
    datos = json.load(f)
    print(json.dumps(datos, indent=2, ensure_ascii=False))
```

### En el siguiente codigo se analizara la disponibilidad de productos por anuncio

```python
disponibilidad = []
for producto, anuncios in datos.items():
    disponibilidad.append((producto, len(anuncios)))

# Ordenar de mayor a menor disponibilidad
disponibilidad.sort(key=lambda x: x[1], reverse=True)

# Mostrar resultados
print("\n Disponibilidad de productos en Revolico:")
for producto, cantidad in disponibilidad:
    print(f" - {producto}: {cantidad} anuncios")
```

### La gráfica muestra la cantidad de anuncios disponibles en Revolico para cada producto analizado, permitiendo comparar cuáles tienen mayor presencia en la plataforma.

```python
import matplotlib.pyplot as plt

# Extraemos productos y cantidades
productos = [producto for producto, cantidad in disponibilidad]
cantidades = [cantidad for producto, cantidad in disponibilidad]

plt.figure(figsize=(12,6))
bars = plt.bar(productos, cantidades, color="#FF9800")

plt.ylabel("Cantidad de anuncios")
plt.xlabel("Productos")
plt.title("Disponibilidad de productos en Revolico")

# Rotar etiquetas del eje X para que no se encimen
plt.xticks(rotation=45, ha="right", fontsize=9)

# Añadir valores encima de cada barra
for bar, cantidad in zip(bars, cantidades):
    plt.text(bar.get_x() + bar.get_width()/2, cantidad + 0.5,
             str(cantidad), ha="center", va="bottom", fontsize=9)

plt.tight_layout()
plt.show()
```

#### El programa organiza los anuncios de cada producto, extrayendo los valores de precio válidos y mostrando la cantidad de ofertas disponibles junto con el precio mínimo, máximo y promedio, lo que permite caracterizar la variabilidad de los costos en el mercado.

```python
def extraer_valor(precio_texto):
    """
    Convierte el campo 'precio_cup' en número si contiene dígitos.
    """
    if not precio_texto:
        return None
    try:
        # Buscar el primer número en el texto
        import re
        match = re.search(r"\d+", precio_texto)
        if match:
            return int(match.group())
        return None
    except:
        return None

print("\n Distribución de precios por producto:")

for producto, anuncios in datos.items():
    precios = []
    for a in anuncios:
        valor = extraer_valor(a.get("precio_cup"))
        if valor is not None:
            precios.append(valor)

    if precios:
        minimo = min(precios)
        maximo = max(precios)
        promedio = sum(precios) / len(precios)

        print(f"\n {producto.upper()}")
        print(f"   Total anuncios : {len(precios)}")
        print(f"   Precio mínimo  : {minimo} CUP")
        print(f"   Precio máximo  : {maximo} CUP")
        print(f"   Precio promedio: {promedio:.0f} CUP")
    else:
        print(f"\n {producto.upper()}")
        print("   No se encontraron precios válidos")
```

### La gráfica muestra la distribución de precios en CUP por producto, destacando los valores mínimos, máximos y promedios obtenidos de los anuncios analizados en Revolico.

```python
import matplotlib.pyplot as plt
import numpy as np

productos = []
minimos = []
maximos = []
promedios = []

for producto, anuncios in datos.items():
    precios = []
    for a in anuncios:
        valor = extraer_valor(a.get("precio_cup"))
        if valor is not None:
            precios.append(valor)
    if precios:
        productos.append(producto)
        minimos.append(min(precios))
        maximos.append(max(precios))
        promedios.append(sum(precios)/len(precios))

# Posiciones
x = np.arange(len(productos))
width = 0.25

plt.figure(figsize=(12,6))
plt.bar(x - width, minimos, width, label="Mínimo", color="#4CAF50")
plt.bar(x, promedios, width, label="Promedio", color="#2196F3")
plt.bar(x + width, maximos, width, label="Máximo", color="#FF9800")

plt.xticks(x, productos, rotation=45, ha="right")
plt.ylabel("Precio en CUP")
plt.title("Precios mínimos, promedios y máximos por producto")
plt.legend()
plt.tight_layout()
plt.show()
```
